```python
# list comprehensions

# normal

num=[]
for i in range(5):
    num.append(i)
print(num)
```

```python
# list comprehension

nums=[i for i in range (5)]
print(nums)
```

```python
# square

square=(x*x for x in range (1,6))
for x in square:
    print(x)
```

```python
# even numbers

even=[x for x in range(26) if x%2==0]
print(even)
```
