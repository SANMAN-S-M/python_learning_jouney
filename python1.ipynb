```python
#operators

# Arithmatic Operators

a=10
b=5
print("Addition: ",a+b)
print("substraction: ",a-b)
print("multipliction: ",a*b)
print("division: ",a/b)
print("modulus: ",a%b)
print("power: ",a**b)
```

```python
# comparison operator

a=10
b=5
print(a==b)
print(a<b)
print(a>b)
print(a<=b)
print(a>=b)
print(a!=b)
```

```python
# Assignment Operators

x=5

x+=3
print(x)

x-=2
print(x)

x*=2
print(x)

x/=2
print(x)
```

```python
# Logical Operators

a=True
b=False

print(a and b)
print(a or b)
print(not a)
```

```python
# Identity Operators

x=[1,2,3]
y=x
z=[1,2,3]

print(x is y)
print(x is z)
print(x is not z)
```

```python
# Membership Operators

name="sanman"

print("s" in name)
print("z" in name)
print("z" not in name)
```

```python
# Bitwise Operators

a=5
b=3
print(a&b)
print(a|b)
print(a^b)
print(~a)
print(a<<b)
print(a>>b)
```

```python
# conditional statement

# IF

age=18

if age>=18:
    print("you are eligible to vote")
    
```

```python
# IF-ELSE

num=5

if num%2==0:
    print("even number")
else:
    print("odd number")
```

```python
#if elif else

marks=75

if marks>=90:
    print("A+")
elif marks>=80:
    print("A")
elif marks>=70:
    print("B+")
else:
    print("fail")
```

```python
# Nested if Statement

age=20
if age>=18:
    if age>=21:
         print("adult and can drive")
    else:
        print("adult")
else:
    print("minor")
```
