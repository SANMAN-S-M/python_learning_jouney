```python
# datatypes

age=20
print(f'age:{age}',type(age))
height=5.9
print(f'height:{height}',type(height))
name='sanman'
print(f'name:{name}',type(name))
is_student=True
print(f'is_student:{is_student}',type(is_student))
marks=[85,90,95]
print(f'marks:{marks}',type(marks))
coordinates=(10,20)
print(f'coordinates:{coordinates}',type(coordinates))
numbers={1,2,3,4,5}
print(f'numbers:{numbers}',type(numbers))
student={'name':'sanman','age':20}
print(f'student:{student}',type(student))
```
